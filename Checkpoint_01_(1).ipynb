# Questão 05 (2 pontos)

Um experimento é realizado para estudar o comportamento de uma mola. Sabe-se que a força aplicada $F$ é proporcional à deformação $x$ da mola, de acordo com o modelo:

$F = kx$,

onde $k$ é a constante elástica da mola.

Alguns pontos experimentais foram coletados e um gráfico foi plotado deles:

```python
x1 = np.array([0., 0.33333333, 0.66666667, 1., 1.33333333, 1.66666667, 2., 2.33333333, 2.66666667, 3.])
y1 = np.array([0.04697955, 1.22843207, 2.05149529, 2.70805268, 4.48304016, 4.98358326, 6.37793699, 7.23133332, 8.33190828, 9.58769491])

# plot
plt.scatter(x1, y1)
plt.show()
```

O seu objetivo é tentar achar qual é o valor da constante da mola ajustando a equação de uma reta para ficar mais adequada.

OBS: Colocando o valor de 'a' no bloco de código abaixo e executando, o gráfico já é gerado automaticamente

# Codigo

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([0., 0.33, 0.66, 1., 1.33, 1.66, 2., 2.33, 2.66, 3.])
y1 = np.array([0.04, 1.22, 2.05, 2.70, 4.48, 4.98, 6.37, 7.23, 8.33, 9.58])

k = 3.17

plt.scatter(x1, y1, label='Dados')
plt.plot(x1, k * x1, 'r', label='F=kx')
plt.legend()
plt.show()
```

Execute essa função para o momento em que você estiver respondendo essa questão.
