```python
<!DOCTYPE html>
<html lang="pt-pt">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Perezbag | Luxury Store</title>
    <script src="https://cdn.tailwindcss.com"></script>
    <link href="https://fonts.googleapis.com/css2?family=Playfair+Display:wght@700&family=Inter:wght@300;400;600&display=swap" rel="stylesheet">
    <script src="https://www.gstatic.com/firebasejs/11.6.1/firebase-app-compat.js"></script>
    <script src="https://www.gstatic.com/firebasejs/11.6.1/firebase-auth-compat.js"></script>
    <script src="https://www.gstatic.com/firebasejs/11.6.1/firebase-firestore-compat.js"></script>
    <style>
        body { font-family: 'Inter', sans-serif; }
        h1, h2, .font-serif { font-family: 'Playfair Display', serif; }
        .gold-text { color: #C5A059; }
        .bg-gold { background-color: #C5A059; }
        .product-card img { transition: transform 0.5s ease; }
        .product-card:hover img { transform: scale(1.05); }
        .hidden-section { display: none; }

        /* Toast Notification Styles */
        .toast {
            position: fixed;
            bottom: 20px;
            right: 20px;
            background: #1A1A1A;
            color: white;
            padding: 15px 25px;
            border-radius: 4px;
            box-shadow: 0 4px 15px rgba(0,0,0,0.2);
            transform: translateY(100px);
            opacity: 0;
            transition: all 0.3s ease;
            z-index: 100;
        }
        .toast.show {
            transform: translateY(0);
            opacity: 1;
        }
    </style>
</head>
<body class="bg-white text-gray-900">

    <!-- NAVIGATION -->
    <nav class="border-b sticky top-0 bg-white z-50 px-6 py-4 flex justify-between items-center">
        <div class="text-2xl font-bold tracking-widest cursor-pointer" onclick="showView('store')">PEREZBAG</div>
        <div class="flex items-center gap-4">
            <button onclick="showView('cart')" class="relative p-2">
                <svg xmlns="http://www.w3.org/2000/svg" class="h-6 w-6" fill="none" viewBox="0 0 24 24" stroke="currentColor">
                    <path stroke-linecap="round" stroke-linejoin="round" stroke-width="2" d="M16 11V7a4 4 0 00-8 0v4M5 9h14l1 12H4L5 9z" />
                </svg>
                <span id="cart-count" class="absolute top-0 right-0 bg-gold text-white text-[10px] rounded-full w-4 h-4 flex items-center justify-center">0</span>
            </button>
            <button onclick="showLoginModal()" class="text-xs uppercase tracking-widest opacity-50 hover:opacity-100">Painel</button>
        </div>
    </nav>

    <!-- STORE FRONT (PUBLIC) -->
    <section id="view-store" class="view-section container mx-auto px-6 py-12">
        <header class="text-center mb-16">
            <p class="subtitle gold-text text-sm tracking-[0.3em] uppercase mb-2">Coleção Exclusiva</p>
            <h2 class="text-4xl md:text-5xl mb-4">O Luxo em Cada Detalhe</h2>
        </header>
        <div id="product-grid" class="grid grid-cols-1 md:grid-cols-2 lg:grid-cols-3 gap-12">
            <!-- Products injected by JS -->
            <div class="col-span-full text-center py-20 opacity-50">Carregando coleção...</div>
        </div>
    </section>

    <!-- CART VIEW -->
    <section id="view-cart" class="view-section hidden-section container mx-auto px-6 py-12">
        <h2 class="text-3xl mb-8 text-center">O Seu Carrinho</h2>
        <div id="cart-items" class="max-w-2xl mx-auto space-y-4 mb-8"></div>
        <div id="cart-total" class="max-w-2xl mx-auto border-t pt-4 text-right mb-8">
            <p class="text-xl font-bold">Total: <span id="total-price">0.00€</span></p>
        </div>
        <div class="flex justify-center gap-4">
            <button onclick="showView('store')" class="px-8 py-3 border border-black uppercase text-xs tracking-widest">Continuar a Comprar</button>
            <button onclick="checkoutWhatsApp()" class="px-8 py-3 bg-black text-white uppercase text-xs tracking-widest">Finalizar no WhatsApp</button>
        </div>
    </section>

    <!-- ADMIN DASHBOARD -->
    <section id="view-admin" class="view-section hidden-section container mx-auto px-6 py-12">
        <div class="flex justify-between items-center mb-10 border-b pb-4">
            <h2 class="text-3xl">Gestão de Inventário</h2>
            <button onclick="logout()" class="text-red-500 text-sm">Sair do Painel</button>
        </div>

        <!-- Add Product Form -->
        <div class="bg-gray-50 p-8 rounded mb-12 shadow-sm border border-gray-100">
            <h3 class="text-xl mb-4 font-serif">Adicionar Nova Bolsa</h3>
            <div class="grid grid-cols-1 md:grid-cols-2 gap-4">
                <div class="flex flex-col gap-1">
                    <label class="text-xs uppercase tracking-widest text-gray-500">Nome do Produto</label>
                    <input type="text" id="p-name" placeholder="ex: Heritage Noir" class="p-3 border rounded focus:outline-none focus:border-black">
                </div>

                <div class="flex flex-col gap-1">
                    <label class="text-xs uppercase tracking-widest text-gray-500">Preço (€)</label>
                    <input type="number" id="p-price" placeholder="ex: 1250" class="p-3 border rounded focus:outline-none focus:border-black">
                </div>

                <div class="flex flex-col gap-1">
                    <label class="text-xs uppercase tracking-widest text-gray-500">Categoria / Etiqueta</label>
                    <select id="p-tag" class="p-3 border rounded focus:outline-none focus:border-black">
                        <option value="Couro Genuíno">Couro Genuíno</option>
                        <option value="Acabamento Manual">Acabamento Manual</option>
                        <option value="Edição Limitada">Edição Limitada</option>
                    </select>
                </div>

                <div class="flex flex-col gap-1">
                    <label class="text-xs uppercase tracking-widest text-gray-500">Foto do Produto (Do seu aparelho)</label>
                    <input type="file" id="p-image-file" accept="image/*" class="p-2 border rounded bg-white cursor-pointer">
                    <p class="text-[10px] text-gray-400 mt-1">A imagem será otimizada automaticamente.</p>
                </div>

                <button id="btn-add-product" onclick="addProduct()" class="md:col-span-2 mt-4 bg-black text-white p-4 uppercase text-sm tracking-widest hover:bg-gold transition">
                    Adicionar ao Catálogo Público
                </button>
            </div>
        </div>

        <h3 class="text-xl mb-6 font-serif">Produtos Atuais no Site</h3>
        <div id="admin-product-list" class="space-y-4"></div>
    </section>

    <!-- CUSTOM CONFIRM MODAL (Deletar Produto) -->
    <div id="confirm-modal" class="fixed inset-0 bg-black/80 hidden z-[70] flex items-center justify-center p-6">
        <div class="bg-white p-8 max-w-sm w-full rounded shadow-2xl text-center">
            <h2 class="text-2xl mb-4 font-serif text-red-600">Remover Produto?</h2>
            <p class="text-gray-600 mb-6 text-sm">Esta ação apagará a bolsa da loja pública permanentemente.</p>
            <div class="flex gap-2">
                <button onclick="closeConfirmModal()" class="flex-1 p-3 border hover:bg-gray-50">Cancelar</button>
                <button id="btn-confirm-delete" class="flex-1 p-3 bg-red-600 text-white hover:bg-red-700">Sim, Remover</button>
            </div>
        </div>
    </div>

    <!-- LOGIN MODAL -->
    <div id="login-modal" class="fixed inset-0 bg-black/80 hidden z-[60] flex items-center justify-center p-6">
        <div class="bg-white p-8 max-w-sm w-full rounded shadow-2xl">
            <h2 class="text-2xl mb-6 font-serif">Acesso Restrito</h2>
            <input type="password" id="admin-pass" placeholder="Senha Master" class="w-full p-3 border mb-4 rounded">
            <div class="flex gap-2">
                <button onclick="closeLoginModal()" class="flex-1 p-3 border">Cancelar</button>
                <button onclick="attemptLogin()" class="flex-1 p-3 bg-black text-white">Entrar</button>
            </div>
            <p id="login-error" class="text-red-500 text-xs mt-3 hidden">Senha incorreta.</p>
        </div>
    </div>

    <!-- FOOTER -->
    <footer class="bg-black text-white py-16 px-6 text-center mt-20">
        <h2 class="text-2xl font-serif mb-4">PEREZBAG</h2>
        <p class="text-gray-500 text-sm max-w-md mx-auto mb-8">Artesanato de excelência e design intemporal para mulheres que valorizam a exclusividade.</p>
        <div class="text-[10px] uppercase tracking-[0.2em] opacity-30">© 2026 Sistema Cloud Perezbag • Permanente</div>
    </footer>

    <!-- TOAST NOTIFICATION CONTAINER -->
    <div id="toast" class="toast">Mensagem</div>

    <script>
        // --- CONFIGURAÇÃO FIREBASE (SITE ETERNO) ---
        // Estas chaves são providenciadas pelo ambiente. O Firebase salva tudo na nuvem.
        const firebaseConfig = JSON.parse(__firebase_config);
        firebase.initializeApp(firebaseConfig);
        const db = firebase.firestore();
        const auth = firebase.auth();
        const appId = typeof __app_id !== 'undefined' ? __app_id : 'perezbag-default';

        // --- ESTADO DA APP ---
        let products = [];
        let cart = [];
        let isAdmin = false;
        let productToDelete = null; // Guarda o ID temporariamente para o modal de confirmação
        const myWhatsApp = "351912345678"; // <--- MUDE PARA O SEU NÚMERO AQUI

        // --- SISTEMA DE NOTIFICAÇÕES (Substitui os alerts feios) ---
        function showToast(message) {
            const toast = document.getElementById('toast');
            toast.innerText = message;
            toast.classList.add('show');
            setTimeout(() => {
                toast.classList.remove('show');
            }, 3000);
        }

        // --- INICIALIZAÇÃO ---
        window.onload = async () => {
            // Login anónimo automático para garantir segurança no Firestore
            if (typeof __initial_auth_token !== 'undefined' && __initial_auth_token) {
                await auth.signInWithCustomToken(__initial_auth_token);
            } else {
                await auth.signInAnonymously();
            }
            loadProducts();
        };

        // --- GESTÃO DE PRODUTOS (Firestore) ---
        function loadProducts() {
            const productRef = db.collection('artifacts').doc(appId).collection('public').doc('data').collection('products');

            productRef.onSnapshot((snapshot) => {
                products = [];
                snapshot.forEach(doc => products.push({id: doc.id, ...doc.data()}));
                renderStore();
                if (isAdmin) renderAdminList();
            }, (err) => console.error("Erro ao carregar nuvem:", err));
        }

        // Função para otimizar a imagem antes de salvar (evita sobrecarregar o banco de dados)
        function compressImage(file) {
            return new Promise((resolve, reject) => {
                const reader = new FileReader();
                reader.readAsDataURL(file);
                reader.onload = (event) => {
                    const img = new Image();
                    img.src = event.target.result;
                    img.onload = () => {
                        const canvas = document.createElement('canvas');
                        const MAX_WIDTH = 600;
                        const MAX_HEIGHT = 800;
                        let width = img.width;
                        let height = img.height;

                        if (width > height) {
                            if (width > MAX_WIDTH) { height *= MAX_WIDTH / width; width = MAX_WIDTH; }
                        } else {
                            if (height > MAX_HEIGHT) { width *= MAX_HEIGHT / height; height = MAX_HEIGHT; }
                        }

                        canvas.width = width;
                        canvas.height = height;
                        const ctx = canvas.getContext('2d');
                        ctx.drawImage(img, 0, 0, width, height);
                        // Comprime para JPEG com 80% de qualidade
                        resolve(canvas.toDataURL('image/jpeg', 0.8));
                    };
                };
                reader.onerror = error => reject(error);
            });
        }

        async function addProduct() {
            const name = document.getElementById('p-name').value;
            const price = document.getElementById('p-price').value;
            const tag = document.getElementById('p-tag').value;
            const fileInput = document.getElementById('p-image-file');

            if(!name || !price) {
                showToast("⚠️ Por favor, preencha o nome e o preço da bolsa.");
                return;
            }

            const btn = document.getElementById('btn-add-product');
            btn.innerText = "A guardar...";
            btn.disabled = true;

            let finalImage = 'https://via.placeholder.com/400x500?text=Perezbag';

            try {
                // Se o utilizador escolheu uma foto, vamos processá-la
                if (fileInput.files && fileInput.files[0]) {
                    finalImage = await compressImage(fileInput.files[0]);
                }

                await db.collection('artifacts').doc(appId).collection('public').doc('data').collection('products').add({
                    name,
                    price: parseFloat(price),
                    tag,
                    image: finalImage,
                    createdAt: new Date().getTime()
                });

                showToast("✅ Produto adicionado com sucesso!");

                // Limpar campos
                document.getElementById('p-name').value = '';
                document.getElementById('p-price').value = '';
                fileInput.value = '';

            } catch (error) {
                showToast("❌ Erro ao guardar. Tente novamente.");
                console.error(error);
            } finally {
                btn.innerText = "Adicionar ao Catálogo Público";
                btn.disabled = false;
            }
        }

        // Funções para o Modal de Confirmação de Eliminação
        function requestDelete(id) {
            productToDelete = id;
            document.getElementById('confirm-modal').classList.remove('hidden');
        }

        function closeConfirmModal() {
            productToDelete = null;
            document.getElementById('confirm-modal').classList.add('hidden');
        }

        document.getElementById('btn-confirm-delete').addEventListener('click', async () => {
            if(productToDelete) {
                await db.collection('artifacts').doc(appId).collection('public').doc('data').collection('products').doc(productToDelete).delete();
                showToast("🗑️ Produto removido da loja.");
                closeConfirmModal();
            }
        });

        // --- RENDERIZAÇÃO ---
        function renderStore() {
            const grid = document.getElementById('product-grid');
            if (products.length === 0) {
                grid.innerHTML = `<div class="col-span-full py-20 text-center opacity-40">Coleção em breve.</div>`;
                return;
            }

            grid.innerHTML = products.map(p => `
                <div class="product-card">
                    <div class="bg-gray-100 aspect-[4/5] overflow-hidden relative mb-4 rounded">
                        <img src="${p.image}" class="w-full h-full object-cover">
                        <span class="absolute top-4 left-4 bg-white/90 px-3 py-1 text-[10px] uppercase font-bold tracking-tighter">${p.tag}</span>
                    </div>
                    <div class="text-center">
                        <h3 class="font-serif text-xl mb-1">${p.name}</h3>
                        <p class="text-gold-text font-semibold mb-4">${p.price.toFixed(2)}€</p>
                        <button onclick="addToCart('${p.id}')" class="w-full py-3 bg-black text-white text-[10px] uppercase tracking-widest hover:bg-gold transition">Adicionar ao Carrinho</button>
                    </div>
                </div>
            `).join('');
        }

        function renderAdminList() {
            const list = document.getElementById('admin-product-list');
            if (products.length === 0) {
                list.innerHTML = `<p class="text-gray-400 text-sm">Nenhum produto cadastrado ainda.</p>`;
                return;
            }

            list.innerHTML = products.map(p => `
                <div class="flex items-center justify-between p-4 bg-white border rounded shadow-sm hover:shadow-md transition">
                    <div class="flex items-center gap-4">
                        <img src="${p.image}" class="w-16 h-16 object-cover rounded border">
                        <div>
                            <p class="font-bold font-serif text-lg">${p.name}</p>
                            <p class="text-sm text-gray-500">${p.price.toFixed(2)}€ &bull; <span class="text-gold-text">${p.tag}</span></p>
                        </div>
                    </div>
                    <button onclick="requestDelete('${p.id}')" class="text-red-500 hover:text-red-700 hover:bg-red-50 px-3 py-1 rounded text-sm transition">Remover</button>
                </div>
            `).join('');
        }

        // --- CARRINHO ---
        function addToCart(id) {
            const product = products.find(p => p.id === id);
            cart.push(product);
            updateCartUI();
            showToast(`👜 ${product.name} adicionada ao carrinho!`);
        }

        function updateCartUI() {
            document.getElementById('cart-count').innerText = cart.length;
            const itemsDiv = document.getElementById('cart-items');
            const totalSpan = document.getElementById('total-price');

            if (cart.length === 0) {
                itemsDiv.innerHTML = `<p class="text-center opacity-50 py-10">O seu carrinho está vazio.</p>`;
                totalSpan.innerText = "0.00€";
                return;
            }

            let total = 0;
            itemsDiv.innerHTML = cart.map((item, index) => {
                total += item.price;
                return `
                    <div class="flex items-center justify-between bg-gray-50 p-4 rounded">
                        <div class="flex items-center gap-4">
                            <img src="${item.image}" class="w-16 h-16 object-cover rounded">
                            <p class="font-serif">${item.name}</p>
                        </div>
                        <div class="flex items-center gap-4">
                            <p>${item.price.toFixed(2)}€</p>
                            <button onclick="removeFromCart(${index})" class="text-gray-400 hover:text-red-500">✕</button>
                        </div>
                    </div>
                `;
            }).join('');
            totalSpan.innerText = total.toFixed(2) + "€";
        }

        function removeFromCart(index) {
            cart.splice(index, 1);
            updateCartUI();
        }

        function checkoutWhatsApp() {
            if (cart.length === 0) {
                showToast("⚠️ O seu carrinho está vazio.");
                return;
            }

            let msg = "✨ *Novo Pedido Perezbag*\n\n";
            let total = 0;
            cart.forEach(item => {
                msg += `👜 ${item.name} - ${item.price.toFixed(2)}€\n`;
                total += item.price;
            });
            msg += `\n💰 *Total: ${total.toFixed(2)}€*\n\n_Por favor, confirme os detalhes de envio._`;

            const url = `https://wa.me/${myWhatsApp}?text=${encodeURIComponent(msg)}`;
            window.open(url, '_blank');
        }

        // --- NAVEGAÇÃO E AUTH ---
        function showView(viewId) {
            document.querySelectorAll('.view-section').forEach(s => s.classList.add('hidden-section'));
            document.getElementById(`view-${viewId}`).classList.remove('hidden-section');
            window.scrollTo(0,0);

            if(viewId === 'admin') {
                renderAdminList();
            }
        }

        function showLoginModal() {
            if(isAdmin) {
                showView('admin');
            } else {
                document.getElementById('login-modal').classList.remove('hidden');
            }
        }

        function closeLoginModal() {
            document.getElementById('login-modal').classList.add('hidden');
        }

        function attemptLogin() {
            const pass = document.getElementById('admin-pass').value;
            if (pass === 'perez123') {
                isAdmin = true;
                closeLoginModal();
                showView('admin');
                document.getElementById('admin-pass').value = '';
                showToast("🔓 Acesso Autorizado");
            } else {
                document.getElementById('login-error').classList.remove('hidden');
            }
        }

        function logout() {
            isAdmin = false;
            showView('store');
            showToast("🔒 Sessão encerrada.");
        }
    </script>
</body>
</html>
```
